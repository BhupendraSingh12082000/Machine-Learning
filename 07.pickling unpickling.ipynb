- ### Import module

```python
import pickle
```

```python
class Employee:
    def __init__(self,eno,ename,esal):
        self.eno = eno
        self.ename = ename
        self.esal = esal

    def display(self):
        print('Employee no is :' , self.eno)
        print('Employee name is :' , self.ename)
        print('Employee salary is :' , self.esal)

#emp = Employee(101,'Bhupendra Singh',15000)
#emp.display()
```

- ### Creating Pickle file

```python
with open('abhi.pkl', 'wb') as f1:
    emp = Employee(101,'Bhupendra Singh',15000)
    pickle.dump(emp,f1)
    print("Pickling is done")
```

- ### Doing Unpickling file

```python
with open('abhi.pkl', 'rb') as f2:
    obj = pickle.load(f2)
    print("Printing Employee Information after unpickling") 
    obj.display()
```

- ### Pickling & Unpickling
    - #### Predict price of a home with area = 5000 sqr ft

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression 
```

```python
# Loading the Dataset
df = pd.read_csv("E:\Data Science Notes\8.MACHINE LEARNING\homeprices\homeprices.csv") 

print(df)

# data Preparation
X = df.drop('price', axis='columns').values
y = df.price.values

# Model creation
reg = LinearRegression() 

# Model training
reg.fit(X, y) 
```

**Pickling**

```python
with open('buddha.pkl','wb') as file1:
    pickle.dump(reg,file1)
```

 **Unpickling**

```python
with open('buddha.pkl','rb') as file2:
    obj = pickle.load(file2)
    print(obj.predict([[5000]]))
```

```python
with open('buddha.pkl','rb') as file2:
    obj = pickle.load(file2)
    print(obj.predict([[3000]]))     # Actual Price 565000
```

```python
with open('buddha.pkl','rb') as file2:
    obj = pickle.load(file2)
    print(obj.predict([[6000]]))    
```

## Saving model, **Joblib & Pickling**

```python
# Loading the Dataset
df = pd.read_csv("E:\Data Science Notes\8.MACHINE LEARNING\homeprices\homeprices.csv") 

print(df)
print()

# data Preparation
X = df.drop('price', axis='columns').values
y = df.price.values

# Model creation
model = LinearRegression() 

# Model training
model.fit(X, y) 

# Pickling
import joblib

joblib.dump(model, 'model_joblib.pkl') 

# Unpickling
mj = joblib.load('model_joblib.pkl') 

print("After Unpickling Predicted Value : ",mj.predict([[5000]]))
print("After Unpickling Predicted Value : ",mj.predict([[6000]]))
print("After Unpickling Predicted Value : ",mj.predict([[3000]]))
```
