- ### Displaying the flower names by using **lambda function**

```python
a = lambda x: iris.target_names[x] 

print(a)
```

```python
df['flower_name'] = df.target.apply(a)

df
```

### Find All **setosa** flowers

## Data Preparation

```python

X = df.drop(['target', 'flower_name'], axis = 'columns')

y = df.target 
```

```python
X.head(3)
```

```python
y.head()
```

## Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X.values , y.values ,test_size = 0.2)
```

## Model creation 

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
model_random = RandomForestClassifier(n_estimators = 40)  #  ( n_estimators = 40 )  parameter defines the number of trees in the random forest. 
```

### Train the model

```python
model_random.fit(X_train,y_train)
```

### Checking the model Score

```python
model_random.score(X_test,y_test)
```

## Model prediction

```python
model_random.predict([[4.8,3.0,1.5,0.3]])
```

```python
model_random.predict([[5.8,2.7,5.1,1.9]])
```
