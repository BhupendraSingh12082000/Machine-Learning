### Import Required Libraries

```python
import pandas as  pd
import numpy as np
```

```python
import warnings 
warnings.filterwarnings('ignore') 
```

### Loading the Dataset

```python
df = pd.read_csv('E:\Data Science Notes\8.MACHINE LEARNING\Melbourne_housing_FULL\Melbourne_housing_FULL.csv') 

df.head()
```

### Find : Rows and columns

```python
df.shape
```

### Find number of Unique values Column wise 

```python
df.nunique()
```

### Accessing only  **required columns**

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 'Distance',
               'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price'] 
 
df = df[cols_to_use] 
```

```python
df.head()
```

```python
df.shape
```

### Checking **NaN** values 

```python
df.isnull().sum()
```

### Few of the columns filling with zero

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car'] 

df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0) 
```

```python
df.isna().sum()
```

### Filling **Landsize & BuildingArea**  columns with mean value

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean()) 
 
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isnull().sum()
```

### Droping NaN Value

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

### Create dummy variabe for categorical data

```python
df = pd.get_dummies(df, drop_first = True) 

df.head()
```

### Data Preparation

```python
X = df.drop('Price', axis = 1) 

y = df['Price'] 
```

### Splitting training and testing datasets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=2) 
```

### Model Creation

```python
from sklearn.linear_model import LinearRegression 
```

```python
reg = LinearRegression()
```

### Train the Model

```python
reg.fit(X_train, y_train)
```

### Cheking the **Train** & **Test** Score

```python
reg.score(X_train, y_train)
```

```python
reg.score(X_test, y_test)*100
```

## Creating **Lasso**  Model Regression

```python
from sklearn.linear_model import Lasso 
```

```python
lasso_reg = Lasso(alpha = 10, max_iter = 10, tol = 0.1)
```

### Model Training with Lasso

```python
lasso_reg.fit(X_train, y_train)
```

### Checking the Score

```python
lasso_reg.score(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

### Prediction

```python
lasso_reg.predict(X_test)
```

### Creating **Ridge** regression model

```python
from sklearn.linear_model import Ridge
```

```python
ridge_reg = Ridge(alpha = 50, max_iter = 100, tol = 0.1)
```

```python
ridge_reg.fit(X_train, y_train)
```

### Checking the Score

```python
ridge_reg.score(X_train, y_train)
```

```python
ridge_reg.score(X_test, y_test)
```

### Prediction

```python
y_test
```

```python
ridge_reg.predict(X_test)
```
